```python
import pandas as pd
import numpy as np
df=pd.read_csv("churn.csv")
```

```python
df.drop("customerID",axis=1,inplace=True)
df["TotalCharges"]=pd.to_numeric(df["TotalCharges"],errors="coerce")
df.dropna(inplace=True)
df["Churn"]=df["Churn"].map({"Yes":1,"No":0})
df_final=pd.get_dummies(df,dtype=int)
```

```python
df_final.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
x=df_final.drop("Churn",axis=1)
y=df_final["Churn"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
n_params={
    "n_estimators":[50,100,200]
}
```

```python
grid_search=GridSearchCV(RandomForestClassifier(random_state=42),n_params,cv=5)
grid_search.fit(x_train,y_train)
```

```python
print(f"best tree :{grid_search.best_params_["n_estimators"]}")
```

```python
y_pred=grid_search.predict(x_test)
```

```python
print(f"دقته مدل:{accuracy_score(y_pred,y_test)*100:.2f}")
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import pandas as pd
best_rf = grid_search.best_estimator_
importances = best_rf.feature_importances_
feature_names = x.columns
feature_importance_df = pd.DataFrame({'ویژگی': feature_names, 'اهمیت': importances}).sort_values(by='اهمیت', ascending=False)
plt.figure(figsize=(10, 8))
plt.barh(feature_importance_df['ویژگی'][:10], feature_importance_df['اهمیت'][:10], color='skyblue')
plt.gca().invert_yaxis()
plt.xlabel('importance')
plt.title('Top 10 Factors Influencing Customer Churn')
plt.show()
```
